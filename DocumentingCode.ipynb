## Finding the functions using the variables

```python
AllVariables=[]
FunctionsUsingOutputs=[]
for Fun in FunctionsVariables:
    AllVariables=AllVariables+Fun[1]
for Fun in FunctionOutputs:    
    AllVariables=AllVariables+Fun[1]
    FunctionsLoc=FunctionsMatDF[Fun[0]]==1
    functions=FunctionsMatDF[FunctionsLoc].index
    for fun in functions:
        FunctionsUsingOutputs.append([fun,Fun[1]])  
Variables=list(set(AllVariables))
Variables.sort()        
FunctionsListDF[Variables]=0
```

```python
Functions_Variables=FunctionsUsingOutputs+FunctionOutputs+FunctionsVariables
```

```python

for variable in Variables:
    Functions=FunctionsListDF[variable]==1
    functions=FunctionsListDF[Functions].index
    for fun in functions:
        print(fun)
        FunctionsLoc=FunctionsMatDF[fun]==1
        print(FunctionsMatDF[FunctionsLoc].index)
```

```python
for Fun in Functions_Variables: #FunctionsVariables: It should include both types of variables
    FunctionName=Fun[0]
    for variable in Fun[1]:
        FunctionsListDF[variable][FunctionName]=1
```

```python
VarFolder=home+'/Variables'
for variable in list(set(AllVariables)):
    FileName=variable+'.md'
    FileLoc=VarFolder+'/'+FileName
    Text = open(FileLoc, "w")
    Text.close()
```

```python
leftside='\n- [['
rightside=']]'
for variable in Variables:
    VariableDocumentationDraft='\n\n### Functions using '+variable+':\n'    
    Functions=FunctionsListDF[variable]==1
    functions=FunctionsListDF[Functions].index
    for fun in functions:
        FunDraft=leftside+fun+rightside
        VariableDocumentationDraft=VariableDocumentationDraft+FunDraft
        FileName=variable+'.md'
    FileLoc=VarFolder+'/'+FileName
    Text = open(FileLoc, "w")
    Text.write(VariableDocumentationDraft)
    Text.close()
```
